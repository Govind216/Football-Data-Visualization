```python
df = pd.read_csv('/content/drive/MyDrive/DVJCOMP/players_22.csv')
df.head()
```

```python
plt.scatter(df['overall'], df['potential'])
```

The scatterplot shows a positive correlation between the two variables, i.e., the points are clustered around a line that slopes upward to the right, it indicates that as a player's current skill level increases, so does their potential skill level (or rating)

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

We see, surprisingly, that among all of a player's attributes, the `movement_reactions` attribute is the one that correlates the most with the overall rating of the player.

```python
sns.regplot(x=df['movement_reactions'], y=df['overall'], scatter_kws={'color': 'lightblue'})
```

This suggests a positive correlation between the 'movement_reactions' and 'overall' variables, i.e., as the 'movement_reactions' score increases, the 'overall' score also tends to increase.
With this piece of information, we will try to build a linear regression model for this

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['movement_reactions'], df['overall'], test_size=0.33)
```

```python
model = LinearRegression()
model.fit(X_train.to_numpy().reshape(-1, 1), y_train)
```

```python
model.score(X_test.to_numpy().reshape(-1, 1), y_test) #calculates the R-squared value of the linear regression model using the testing data.
```

```python
from sklearn.svm import SVR
```

```python
model = SVR()
model.fit(X_train.to_numpy().reshape(-1, 1), y_train)
model.score(X_test.to_numpy().reshape(-1, 1), y_test)
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
model = Lasso()
model.fit(X_train.to_numpy().reshape(-1, 1), y_train)
model.score(X_test.to_numpy().reshape(-1, 1), y_test)
```

```python
model = Ridge()
model.fit(X_train.to_numpy().reshape(-1, 1), y_train)
model.score(X_test.to_numpy().reshape(-1, 1), y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
model = DecisionTreeRegressor()
model.fit(X_train.to_numpy().reshape(-1, 1), y_train)
model.score(X_test.to_numpy().reshape(-1, 1), y_test)
```

```python
df['player_positions_list'] = df.player_positions.str.split(',')
df['player_positions_list']
```

```python
df_exploded = df.explode('player_positions_list')
df_exploded.head()
```

```python
attributes = df_exploded.columns[37:77]
```

```python
attributes
```

```python
mask = df_exploded[attributes].isna().any(axis=1)
```

```python
X = df_exploded[attributes][~mask]
y = df_exploded['player_positions_list'][~mask]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
_ = plot_tree(model)
```

```python
mentalities = list(filter(lambda x: x.startswith('mentality'), attributes))
```

```python
fig, ax = plt.subplots(nrows=3, ncols=2)
i = 0
for row in ax:
  for col in row:
    col.plot(df[mentalities[i]], df.overall)
    i += 1
```

```python
sns.heatmap(df[mentalities + ['overall']].corr(), annot=True)
```

From this we can see that For Overall rating, composure correlates the most, thus a player that is composed has a higher likelihood of being a better player and coaches should look for this while scouting.

We also see that players with better positioning are the better penalty takers.

We also see that players with aggressive mentalities have a better interception rate, so in the case of midfielders it might be better to select player with an aggressive mentality
